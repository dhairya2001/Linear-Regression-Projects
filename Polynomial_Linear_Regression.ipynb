Import Important Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Position_Salaries.csv')
print(dataset)
```

```python
X=dataset.iloc[:,1:2].values #reshaping in 2D matrix
Y=dataset.iloc[:,2].values
```

NO training and test set as data is small and we need accurate model

```python
#fitting Linear regression to dataset
from sklearn.linear_model import LinearRegression
model_1 = LinearRegression()
model_1.fit(X,Y)
```

```python
#fitting polynomial regression to dataset
from sklearn.preprocessing import PolynomialFeatures
poly_model = PolynomialFeatures(degree=2) #default degree is also 2 
X_poly = poly_model.fit_transform(X)
poly_model.fit(X_poly,Y)
```

```python
print(X_poly)
```

First Column is constant , second is variable x like in dataset, third is polynomial variable

```python
model_2=LinearRegression()
model_2.fit(X_poly,Y)
```

```python
#visualilsing linear regression result
plt.scatter(X,Y,color='green')
plt.plot(X, model_1.predict(X),color='red')
plt.title('Linear Regression Result')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

It is clearly seen it is not accurate prediction, green points are actual salary and red line is predicted salary

```python
#visulaising Polynomial regression result (degree=2)
plt.scatter(X,Y,color='green')
plt.plot(X, model_2.predict(poly_model.fit_transform(X)),color='red')
plt.title('Polynomial Regression Result')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

```python
#fitting polynomial regression to dataset for degree=3
from sklearn.preprocessing import PolynomialFeatures
poly_model = PolynomialFeatures(degree=3)
X_poly = poly_model.fit_transform(X)
poly_model.fit(X_poly,Y)
model_2=LinearRegression()
model_2.fit(X_poly,Y)
```

```python
#visulaising Polynomial regression result (degree=3)
plt.scatter(X,Y,color='green')
plt.plot(X, model_2.predict(poly_model.fit_transform(X)),color='red')
plt.title('Polynomial Regression Result')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

```python
#fitting polynomial regression to dataset for degree=4
from sklearn.preprocessing import PolynomialFeatures
poly_model = PolynomialFeatures(degree=4)
X_poly = poly_model.fit_transform(X)
poly_model.fit(X_poly,Y)
model_2=LinearRegression()
model_2.fit(X_poly,Y)
```

```python
#visulaising Polynomial regression result (degree=4)
plt.scatter(X,Y,color='green')
plt.plot(X, model_2.predict(poly_model.fit_transform(X)),color='red')
plt.title('Polynomial Regression Result')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

For degree=4 the graph seems extremely accurate as compared to degree-2,3

```python
#Predicting a new result with Linear Regression for a level of 6.5
level=[[6.5]]
model_1.predict(level)
```

```python
#Predicting a new result with Polynomial Regression for a level of 6.5
model_2.predict(poly_model.fit_transform(level))
```

Polynomial Regression is more accurate as compared to Linear Regression in this case
